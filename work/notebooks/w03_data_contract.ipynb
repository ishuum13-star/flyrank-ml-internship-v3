## 1. Unit of analysis + time window

*One row = one what, over which dates? State it, then verify it below.*

One row = one content page (content_hash_id) belonging to one client (client_hash_id),
aggregated over report_date within a single calendar month. Time window: month=2026-03
(March 2026) — a mid-panel month, not the final month (June 2026 is reserved as the
sealed test month, per the warehouse's own _sample file). The raw fact_content_daily_performance
table is daily-grain (one row = one content page-day); I aggregate it to page-month grain
(sum impressions/clicks, average position) before building my lane's feature frame.

## 2. Fields: feature / label / context / excluded

*Sort every field you plan to touch into these four buckets. Excluded needs a why.*

FEATURE: dim_content.content_type, main_intent, word_count, char_count, search_volume,
competition, cpc, backlinks; aggregated from fact — sum(gsc_impressions), sum(gsc_clicks),
avg(gsc_avg_position) for March.

LABEL/PROXY: declining = 1 if sum(gsc_impressions) in March 2026 < sum(gsc_impressions) in
February 2026, for the same content_hash_id. A rule-derived proxy, not a directly observed
"this page failed" outcome — it depends on GSC impressions being tracked consistently both months.

CONTEXT (used for joins/filters, not fed as a feature): report_date, client_hash_id,
content_hash_id, month, content_created_date, last_optimized_date, client_has_gsc/ga4_data_available.

EXCLUDED: content_hash_id / client_hash_id / keyword_hash_id / url_hash_id — salted hash
identifiers; excluding them as features stops the model from memorizing specific pages
instead of learning a generalizable pattern. Also excluded: provider_used / model_used —
metadata about how content was produced, not something I want driving prioritization decisions
this early. is_deleted — a post-hoc status not knowable at the decision moment for a live page.

## 3. Verify it with queries (grain, counts, missing values, windows)

*Every claim above gets a query cell here. A contract claim without a query next to it is a guess.*

Query A — grain check:

```python
grain_check = con.sql(f"""
    SELECT content_hash_id, report_date, COUNT(*) AS n
    FROM '{BASE}/fact_content_daily_performance/month=2026-03/data_0.parquet'
    GROUP BY content_hash_id, report_date
    HAVING COUNT(*) > 1
    LIMIT 5
""").df()
print("Rows where (content_hash_id, report_date) is NOT unique:", len(grain_check))
grain_check
```

Query B — row count + date span:

```python
span = con.sql(f"""
    SELECT COUNT(*) AS n_rows,
           COUNT(DISTINCT content_hash_id) AS n_pages,
           MIN(report_date) AS min_date,
           MAX(report_date) AS max_date
    FROM '{BASE}/fact_content_daily_performance/month=2026-03/data_0.parquet'
""").df()
span
```

Query C — availability, filtered with IS TRUE:

```python
avail = con.sql(f"""
    SELECT COUNT(*) AS total_rows,
           COUNT(*) FILTER (WHERE gsc_data_available IS TRUE) AS gsc_available_rows
    FROM '{BASE}/fact_content_daily_performance/month=2026-03/data_0.parquet'
""").df()
avail
```

February + March data aggregate

```python
feb = con.sql(f"""
    SELECT content_hash_id, SUM(gsc_impressions) AS impressions_feb
    FROM '{BASE}/fact_content_daily_performance/month=2026-02/data_0.parquet'
    GROUP BY content_hash_id
""").df()

mar = con.sql(f"""
    SELECT content_hash_id, SUM(gsc_impressions) AS impressions_mar
    FROM '{BASE}/fact_content_daily_performance/month=2026-03/data_0.parquet'
    GROUP BY content_hash_id
""").df()

content_attrs = con.sql(f"""
    SELECT content_hash_id, word_count, content_type, search_volume, competition
    FROM '{BASE}/dim_content.parquet'
""").df()

print("Feb pages:", len(feb), "| March pages:", len(mar), "| Content attrs:", len(content_attrs))
```

```python
# Join Feb (features) + March (label source) + content attributes
lane = feb.merge(mar, on="content_hash_id", how="inner") \
          .merge(content_attrs, on="content_hash_id", how="inner")

# Label/proxy: declining if March impressions dropped vs February
lane["declining"] = (lane["impressions_mar"] < lane["impressions_feb"]).astype(int)

print("Lane rows after join:", len(lane))
print("Declining rate:", round(lane["declining"].mean(), 3))
lane[["content_hash_id", "impressions_feb", "impressions_mar", "word_count",
      "content_type", "search_volume", "competition", "declining"]].head()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_auc_score

# Honest features — all knowable BEFORE March happens (from Feb data + static content attrs)
honest_features = ["impressions_feb", "word_count", "search_volume", "competition"]
X_honest = lane[honest_features].fillna(0)
y = lane["declining"]

tree_honest = DecisionTreeClassifier(max_depth=3, class_weight="balanced", random_state=42)
tree_honest.fit(X_honest, y)
score_honest = roc_auc_score(y, tree_honest.predict_proba(X_honest)[:, 1])
print(f"Honest baseline ROC AUC: {score_honest:.3f}")
```

```python
# THE TRAP — add a feature computed FROM March data, which is exactly what we're predicting
lane["pct_change_leak"] = (lane["impressions_mar"] - lane["impressions_feb"]) / lane["impressions_feb"].replace(0, 1)

leaky_features = honest_features + ["pct_change_leak"]
X_leaky = lane[leaky_features].fillna(0)

tree_leaky = DecisionTreeClassifier(max_depth=3, class_weight="balanced", random_state=42)
tree_leaky.fit(X_leaky, y)
score_leaky = roc_auc_score(y, tree_leaky.predict_proba(X_leaky)[:, 1])
print(f"'Leaky' ROC AUC: {score_leaky:.3f}  <- looks amazing, but it's cheating")
print(f"\nWhy: pct_change_leak is calculated directly from impressions_mar — the same")
print(f"March data 'declining' is derived from. It's not a real feature, it's the answer in disguise.")
```

```python
# Remove the leak — this is the number that actually means something
lane = lane.drop(columns=["pct_change_leak"])
print(f"Kept: honest baseline ROC AUC = {score_honest:.3f} (not the fake {score_leaky:.3f})")
```

Five features (max), each knowable before the decision moment (before March 2026 starts):

1. impressions_feb — knowable at decision moment because it's fully observed by end of
   February, before we decide which pages to prioritize for March.
2. word_count — knowable because it's a static content attribute set when the page was
   published, unrelated to any future performance.
3. search_volume — knowable because it's a keyword-level metric independent of this
   page's actual performance.
4. competition — same as search_volume: a keyword attribute, not derived from outcomes.
5. content_type — knowable because it's fixed at publish time (e.g. "keyword article"),
   not something that changes based on performance.

Honest baseline (these 4 numeric features + content_type, via a depth-3 tree): ROC AUC 0.849.

## 4. Data limits

*What can this data never tell you? Unbalanced history, GSC-only early rows, window overlaps.*

This data can never tell you: (1) WHY a page declined — only that GSC impressions dropped,
not the cause (algorithm change, seasonality, a competitor, a technical issue). (2) The full
picture for every page — only 36.7% of March rows have gsc_data_available IS TRUE; the other
63% are missing GSC data for that day, which could bias any feature built from daily granularity
rather than the monthly sum used here. (3) Client history evenly — dim_clients shows an
unbalanced panel (gsc_data_start/ga4_data_start differ per client), so newer clients have
less history than older ones, and any model trained on this slice will underrepresent
newer clients. (4) Causation — a declining label is an observed correlation with time,
not proof that anything about the page itself caused the drop.

```python
# Backing the "data limits" claims with numbers
print(f"GSC availability in March: {3611061/9841378:.1%} of rows")

unbalanced = con.sql(f"SELECT gsc_data_start, COUNT(*) as n_clients FROM '{BASE}/dim_clients.parquet' GROUP BY gsc_data_start ORDER BY n_clients DESC LIMIT 5").df()
print("\nClient history start dates vary (unbalanced panel):")
unbalanced
```
